```python
# Importação de bibliotecas necessárias
import google.generativeai as genai
from google.colab import userdata
from IPython.display import display, Markdown, Image, clear_output
import re
import ipywidgets as widgets

# Importação de módulos personalizados

# Configuração da chave de API do Google Generative AI
api_key = userdata.get('SECRET_KEY')
genai.configure(api_key=api_key)

# Configurações de geração de texto e segurança
generation_config = {
    "candidate_count": 1,
    "temperature": 0.3,
}

safety_settings = {
    "HARASSMENT": "BLOCK_NONE",
    "HATE": "BLOCK_NONE",
    "SEXUAL": "BLOCK_NONE",
    "DANGEROUS": "BLOCK_NONE",
}

# Carregamento de imagens
IMAGEM_URL = "https://raw.githubusercontent.com/AGZero17/My_First_AIProject/main/modelo.PNG"  #
IMAGEM = Image(url=IMAGEM_URL)



# Inicialização de modelos de geração de texto
model_rotina = genai.GenerativeModel(model_name='gemini-1.0-pro', generation_config=generation_config, safety_settings=safety_settings)

# Lista para armazenar as respostas
respostas = []

def filtro_respostas(pergunta, resposta):
    """
    Esta função filtra as respostas do usuário com base na pergunta feita.

    Argumentos:
    pergunta (str): A pergunta feita ao usuário.
    resposta (str): A resposta fornecida pelo usuário.

    Retorna:
    resposta_filtrada: A resposta filtrada ou None se for inválida.
    """
    # Checagem da idade
    if pergunta.startswith('Qual a sua idade'):
        try:
            # Convertendo a resposta para um número inteiro
            idade = int(re.sub(r'\D', '', resposta))  # Removendo caracteres não numéricos
            if 10 <= idade <= 120:  # Verificando se a idade está dentro do intervalo válido
                return idade
            else:
                print("Idade inválida. Informe uma idade entre 1 e 120 anos.")
                return None
        except ValueError:
            print("Por favor, informe uma idade válida!")
            return None

    # Checagem do peso
    elif pergunta.startswith('Qual é o seu peso'):
        try:
            # Convertendo a resposta para um número inteiro
            return int(resposta)
        except ValueError:
            print('Por favor, informe um peso válido!')
            return None

    # Checagem da altura
    elif pergunta.startswith('Qual é a sua altura'):
        try:
            altura = float(re.sub(r'\D', '', resposta))  # Removendo caracteres não numéricos
            if altura <= 240:  # Verificando se a altura é válida
                altura_str = str(altura)
                if len(altura_str) == 3:
                     altura_str = altura_str[:1] + '.' + altura_str[1:]  # Formatando altura para metros
                     altura = float(altura_str)
                return altura
            else:
                print("Altura inválida. A altura não pode ser maior que 2.40 metros.")
                return None
        except ValueError:
            print('Por favor, informe uma altura válida!')
            return None

    # Checagem da avaliação da alimentação diária
    elif pergunta.startswith('Como você avalia sua alimentação diária'):
        opcoes = ['muito boa', 'boa', 'normal', 'ruim', 'muito ruim']
        if resposta.lower() not in opcoes:
            print("Por favor, escolha uma das opções abaixo!")
            return None
        return resposta.lower()

    # Se não for uma pergunta específica, retorna a resposta sem filtragem
    else:
        return resposta

# Introdução ao usuário
display(Markdown(f"**Olá prazer em conhecelo meu nome é NutriTrAIn! A partir de agora serei seu instrutor pessoal de exercícios e nutrição personalizado, meu objetivo é ajudá-lo a atingir seus objetivos de saúde e condicionamento físico de forma eficaz e sustentável.**"))
display(Markdown(f"**Irei realizar um teste logo abaixo, para eu conhece-lo melhor e assim irei montar o seu programa!**💪"))
print()
display(Markdown(f"**❗Atenção❗ seja sincero ao responder as perguntas ❤**"))
print('-='*60)
# Função para exibir a próxima pergunta
def exibir_proxima_pergunta(i=0):
    if i < len(perguntas):
        pergunta = perguntas[i]
        # Centralizando o texto das perguntas e definindo cor de fundo
        label = widgets.Label(value=pergunta, style={'description_width': 'initial'},
                             layout=widgets.Layout(justify_content='center', background_color='lightblue'))
        text_input = widgets.Text(layout=widgets.Layout(width='500px'))
        display(label)
        display(text_input)

        def on_submit(sender):
            resposta = filtro_respostas(pergunta, text_input.value)
            if resposta is not None:
                respostas.append(resposta)
                text_input.disabled = True  # Desabilita o input após a resposta
                exibir_proxima_pergunta(i + 1)
            else:
                print("Por favor, preencha a pergunta corretamente.")

        text_input.on_submit(on_submit)
    else:
        mostrar_resultados(respostas)

# Função para mostrar os resultados (chamada após a última pergunta)
def mostrar_resultados(respostas):
    # Calcula o IMC
    IMC = int(respostas[3]) / (float(str(respostas[4]).replace(",", ".")) ** 2)

    # Cria o texto generativo com as informações do usuário
    texto_generativo = (
        "-> Atue como um instrutor profissional com expertise em:\n"
        "   - Medicina do Esporte\n"
        '   - Personal Training (PRINCIPAL)\n'
        '   - Nutrição Esportiva (PRINCIPAL)\n'
        '   - Ortopedia\n'
        '   - Fisioterapia\n'
        '   - Endocrinologia\n'
        '   - Cardiologia (PRINCIPAL)\n'
        "----------------------------------------------------------------------------------------"

        f'   Qual o seu nome? {respostas[0]}\n',
        f'   Qual o seu sexo? {respostas[1]}\n',
        f'   Qual a sua idade? {respostas[2]}\n',
        f'   Qual é o seu peso atual? {respostas[3]}\n',
        f'   Qual é a sua altura? {respostas[4]}\n',
        f'   Você toma medicamento(s)? Qual(is)? {respostas[5]}\n',
        f'   Você já teve algum problema de saúde cardiovascular? {respostas[6]}\n',
        f'   Você já teve algum problema de saúde relacionado a lesões musculares ou articulares durante algum treino ou exercício? {respostas[7]}\n',
        f'   Você costuma praticar exercícios físicos regularmente? Se sim, sente fadiga excessiva ao treinar? {respostas[8]}\n',
        f'   Quantos copos de água você toma por dia? {respostas[9]}\n',
        f'   Você já consultou algum profissional de saúde anteriormente para avaliar sua condição física em relação aos exercícios? {respostas[10]}\n',
        f'   Pratica esporte(s)? Se sim, qual(is)? {respostas[11]}\n',
        f'   Você tem equipamentos de treino (pesos, barra, ...)? Se sim, qual(is)? {respostas[12]}\n',
        f'   Com que frequência você consome alimentos processados ou ricos em açúcar? {respostas[13]}\n',
        f'   Quantas horas você dorme por noite? {respostas[14]}\n',
        f'   Como você descreveria seu nível de estresse no dia a dia? {respostas[15]}\n',
        f'   Você tem algum histórico familiar de doenças crônicas, como diabetes ou hipertensão? {respostas[16]}\n',
        f'   O que te motiva a querer melhorar sua saúde e condicionamento físico? {respostas[17]}\n'
        f'   O IMC do usuario é: {IMC}\n'
        "----------------------------------------------------------------------------------------"
        '   Desenvolva rotinas de exercícios e alimentação para promover a saúde, ajustando conforme necessário para evitar complicações devido a doenças. Destaque o design com emojis.\n\n'
        "----------------------------------------------------------------------------------------"
        'CRITÉRIOS:\n'
        '- Descreva os exercícios de força (flexões, abdominais, prancha, etc.) sem depender de equipamentos.\n'
        '- Aprimore o design com emojis, enfocando títulos, subtítulos e dieta.\n'
        '- Adapte os exercícios de acordo com a idade e capacidade do usuário.\n'
        '- Ofereça instruções claras para exercícios complexos.\n'
        '- Para IMC ≥ 25, reduza o número de exercícios para evitar sobrecarga.\n'
        '- Organize a rotina por dia da semana, detalhando exercícios, descanso e tempo de treino em uma tabela.\n'
        '- Personalize a rotina com base nos medicamentos do usuário, evitando impactos negativos na saúde.\n\n'
        "----------------------------------------------------------------------------------------"
        'EXEMPLO:\n'
        '|Nome: João|\n',
        '|Sexo: Masculino|\n',
        '|Idade: 30 anos|\n',
        '|Peso: 80 Kg|\n',
        '|Altura: 1.75 m|\n',
        '|Remédio(s): Nenhum|\n',
        '|Exercícios regulares? Sim, pratico 3 vezes por semana. Não sinto fadiga excessiva.|\n',
        '|Avaliação da alimentação: Boa, mas consumo alimentos processados ocasionalmente.|\n',
        '|Problema cardiovascular? Não|\n',
        '|Lesões musculares/articulares? Não|\n',
        '|Copos de água por dia: Cerca de 8 copos|\n',
        '|Consultou profissional de saúde? Sim, consultei um médico há 6 meses atrás|\n',
        '|IMC: 26.1|\n',
        '|Esporte praticado: Futebol recreativo aos finais de semana|\n',
        '|Equipamentos de treino? Sim, possuo halteres e uma esteira|\n',
        '|Meta de saúde: Perder peso e ganhar condicionamento físico|\n',
        '|Com que frequência você consome alimentos processados ou ricos em açúcar? Ocasionalmente, algumas vezes por semana|\n',
        '|Quantas horas você dorme por noite? Cerca de 7 a 8 horas|\n',
        '|Como você descreveria seu nível de estresse no dia a dia? Moderado, às vezes me sinto um pouco estressado com o trabalho|\n',
        '|Você tem algum histórico familiar de doenças crônicas, como diabetes ou hipertensão? Não|\n',
        '|O que te motiva a querer melhorar sua saúde e condicionamento físico? Quero ter mais energia e me sentir melhor no dia a dia|\n'
        "----------------------------------------------------------------------------------------"
        'Rotina de Exercícios Personalizada:\n'
        '- Segunda-feira:\n'
        '  - Aeróbico: Caminhada rápida ou corrida leve por 30 minutos.\n'
        '  - Força: Agachamentos (3x12), Flexões de braço (3x10), Abdominais (3x15).\n'
        '- Terça-feira:\n'
        '  - Flexibilidade: Alongamento completo por 15 minutos.\n'
        '- Quarta-feira:\n'
        '  - Descanso Ativo: Atividades leves, como yoga, por 30 minutos.\n'
        '- Quinta-feira:\n'
        '  - Aeróbico: Caminhada rápida ou corrida leve por 30 minutos.\n'
        '  - Força: Levantamento de peso (3x10), Flexões modificadas (3x8), Prancha (3x30s).\n'
        '- Sexta-feira:\n'
        '  - Flexibilidade: Alongamento específico por 20 minutos.\n'
        '- Sábado:\n'
        '  - Aeróbico: Escolha uma atividade por 45 minutos.\n'
        '- Domingo:\n'
        '  - Descanso Ativo: Passeio leve por 30 minutos.\n\n'
        'Observações:\n'
        '- Faça aquecimento antes dos exercícios.\n'
        '- Mantenha-se hidratado.\n'
        '- Ajuste a intensidade conforme necessário.\n'
        '- Consulte um profissional de saúde se estiver tomando medicamentos.\n\n'

        '-------------------------------------------------------------------------------------------------------'
        "-------------------------------------------------------------------------------------------------------"
        "CRITÉRIOS PARA CRIAR A ROTINA ALIMENTAR DO USUÁRIO"
        "-------------------------------------------------------------------------------------------------"

        "- Preste atenção na idade, sexo, peso, altura e nos remédios do usuário para criar a rotina, com base neles"
        "- Procure colocar valores de xícara e colher APENAS EM alimentos que possam ser liquidos, lacticíneos e grãos (1/2 de açucar, 1 xícara de chá, 1 colher de requeijão, ...) "
        "- EVITE COLOCAR VALORES DE XÍCARA EM ALIMENTOS COMPLETOS (SOPA, CARNE, PEIXES) E QUE NÃO SÃO DE COLOCAR EM XÍCARAS "
        "- Coloque valores de peso (g - gramas, kg - quilogramas) em alimentos derivados de animais, grãos (carne, peixe, frango, açucar, quinoa) "
        "- Com base na resposta dos remédios, procure para oque ele é usado, logo sabendo o porquê do usuário usa altere a rotina colocando exercícios que não comprometam com a saúde do usuário ou que cause consequências caso ele tenha alguma doença em forma de tabela\n"
        f"- Agora é sua vez, crie uma rotina alimentar baseada nestes exemplos"
        "- De dicas ao usuário para administrar sua rotina"
        "-------------------------------------------------------------------------------------------------"
        "EXEMPLO ABAIXO:"
        "-------------------------------------------------------------------------------------------------"

          "Rotina Alimentar Recomendada:"

          "*Café da Manhã:*"
          "  - [Alimento 1]"
          "  - [Alimento 2]"
          "  - [Alimento 3]"
          "*Lanche da Manhã:*"
          "  - [Alimento 4]"
          "  - [Alimento 5]"
          "*Almoço:*"
          "  - [Alimento 6]"
          "  - [Alimento 7]"
          "  - [Alimento 8]"
          "*Lanche da Tarde:*"
          "  - [Alimento 9]"
          "  - [Alimento 10]"
          "*Jantar:*"
          "  - [Alimento 11]"
          "  - [Alimento 12]"

        "-------------------------------------------------------------------------------------------------"

        "**Evite Excessos:**"

        "- Reduza a ingestão de petiscos e alimentos processados."
        "- Mantenha-se hidratado, bebendo bastante água ao longo do dia."
        "- Evite refeições pesadas antes dos treinos para evitar desconforto durante o exercício."
        "-------------------------------------------------------------------------------------------------"


        "É importante manter uma dieta balanceada e variada, incluindo alimentos de todos os grupos alimentares. Além disso, lembre-se de ajustar as porções de acordo com suas necessidades calóricas e objetivos de saúde."

        "Se tiver alguma dúvida ou precisar de mais orientações, estou à disposição para ajudar. Lembre-se sempre de consultar um profissional de saúde antes de fazer mudanças significativas em sua dieta."

        "Atenciosamente,"
        f"NutriTrAIn"

        "-------------------------------------------------------------------------------------------------"
        f'Agora retorne a rotina de exercícios personalizada com as informações coletadas de {respostas}'
        "-------------------------------------------------------------------------------------------------"
        "-------------------------------------------------------------------------------------------------"
        f"E a rotina alimentar com base nos seguintes alimentos {alimentos_saudaveis}, de forma coerente, evitando informações sem sentindo (1 colher de carne, 1/2 xícara de salmão, 1/4 xícara de salada e etc...) :"
    )

    print()
    print("-=" * 60)
    # Exibe uma mensagem personalizada para o usuário
    display(Markdown(f"**{respostas[0]}, Com base nas suas informações e objetivo de saúde elaborei sua rotina de exercícios e alimentação, espero que aproveite!**"))
    print("-=" * 60)
    display(Markdown(f'**Apenas aguarde alguns instantes..**'))
    # Gera as rotinas usando generate_content
    response = model_rotina.generate_content(texto_generativo)
    display(Markdown(response.text))

    print("-=" * 60)

   # Widget de botão e entrada de texto para ajuda adicional
    botao_ajuda = widgets.Button(description="Quer ajuda?", style={'button_width': '350px', 'button_height': '50px'})
    input_ajuda = widgets.Text(placeholder="-> Digite sua dúvida aqui", style={'button_width': '200px', 'button_height': '50px'})
    print("-=" * 60)

    output_ajuda = widgets.Output()

    def on_input_submit(sender):
        with output_ajuda:
            clear_output()
            duvida = input_ajuda.value
            novo_prompt = f"{texto_generativo} \n\nO usuário precisa de ajuda com: {duvida}\n\nPor favor, forneça mais informações sobre: {duvida}."
            ajuda = model_rotina.generate_content(novo_prompt)
            print('-='*60)
            display(Markdown(ajuda.text))

    # Associa a função on_input_submit ao evento on_submit do input_ajuda
    input_ajuda.on_submit(on_input_submit)

    display(widgets.VBox([botao_ajuda, input_ajuda, output_ajuda]))

# Inicia o processo exibindo a primeira pergunta
exibir_proxima_pergunta()
```
